```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
obj = '162p'
main_dir = '/Users/s1523386/Documents/year1/shape_modelling/'
datadir = main_dir+obj+'/period_scan/'
file = datadir+'out_periods_'+obj

period = []
chi = []

f = open(file)

lines = f.readlines()

for line in lines:
    data = line.split()
    period.append(float(data[0]))
    chi.append(float(data[2]))
```

```python
np.shape(period)
```

```python
for i in range(len(chi)):
    val = chi[i]
    if val ==min(chi):
        no = i
print('Min chi^2, period @ min chi^2 =', chi[no], period[no])
```

```python
for i in range (len(chi)):
    print(period[i], chi[i])
```

```python
chi_sorted = np.sort(chi)
for i,val in enumerate(chi):
    if chi[i]==chi_sorted[1]:
        period2=period[i]
        print(min(chi),chi[i],period2)
```
