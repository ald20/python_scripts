```python
# Create script to test different values of period range to sample between for period scan
# Aim for n~100

import numpy as np
import os
```

```python
def calc_n(p1,p2,t):
    # p1 = lower period interval
    # p2 = upper period interval
    # t = delta-t = range in JD (days) between first and last observation
    # p = 0.8 (Scaling parameter - best left unchanged)
    p = 0.8;
    t_hours = t*24.    # Convert t into hours
    
    n = (2*t_hours*(p2-p1))/(p*p1**2)
    return n
```

```python
# Define t:
t = jd_list[-1]-jd_list[0]
print("t = ", '% .6f' %t, 'days')
```

```python
# Read in synodic rotation period from literature:
p_lit = 32.853    # hours

# Define min and max periods such that p_lit is centred between these:
p_lit_min = p_lit-2.853
p_lit_max = p_lit+1.147
#p_lit_min=31
#p_lit_max=34
print('Want to perform period scan between % .4f' %p_lit_min, ' and % .4f' %p_lit_max, 'hours')
```

```python
# Increment period by range_test to see how large n would be 
range_test = 0.35


print('Range test of step size %.4f would mean n=%d steps'%(range_test,calc_n(p_lit_min, p_lit_min+range_test,t)))
```

```python
period=p_lit_min

# Make array of period ranges to test between
period_range = []
while period<=p_lit_max:
    print('% .4f' %period)
    period = period+range_test
    period_range.append((round(period-range_test,4), round(period, 4)))
```

```python
period_range = np.array(period_range[:-1])
period_range
```

```python
def make_input_period(p1,p2,p):
    return "%.4f %.4f %.1f period start - end - interval coeff.\n0.1    convexity weight\n6 6     degree and order of spherical harmonics\n8       no. of rows\n0 0   scattering parameters\n0 0 \n-0.5  0\n0.1 0\n300       iteration stop condition\n10      minimum number of iterations (only if the above value < 1)\n"%(p1, p2, p)
```

```python
dir_name = main_dir+'period_scan'
if not os.path.exists(dir_name):
    os.makedirs(dir_name)

for i in range(1,len(period_range)+1):
    if i<10:
        #print('00%i'%i)
        fout = open("%s/input_period_00%i"%(dir_name,i),'w+')
        fout.write(make_input_period(period_range[i-1,0], period_range[i-1,1], 0.8))
        fout.close()
    else:
        #print('0%i'%i)
        fout = open("%s/input_period_0%i"%(dir_name,i),'w+')
        fout.write(make_input_period(period_range[i-1,0], period_range[i-1,1], 0.8))
        fout.close()    
#for i in range(13,len(period_range)+13):
#    #print('0%i'%i)
#    stri=i-12
#    fout = open("%s/input_period_0%i"%(dir_name,i), 'w+')
#    fout.write(make_input_period(period_range[stri-1,0], period_range[stri-1,1], 0.8))
#    fout.close()
```
